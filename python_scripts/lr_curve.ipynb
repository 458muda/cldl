```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def numpy_isqrt(number):
    threehalfs = 1.5
    x2 = number * 0.5
    y = np.float32(number)
    
    i = y.view(np.int32)
    i = np.int32(0x5f3759df) - np.int32(i >> 1)
    y = i.view(np.float32)
    
    y = y * (threehalfs - (x2 * y * y))
    return float(y)


def scheduler(epoch):
    d_model = 1000000
    warmup_steps = 50

    arg1 = numpy_isqrt(epoch)
    arg2 = epoch * (warmup_steps ** -1.5)

    return numpy_isqrt(d_model) * min(arg1, arg2)

```

```python
x= []
y = []

for a in range(1000):
    x.append(a)
    y.append(scheduler(a))
```

```python
plt.plot(x,y)
```
